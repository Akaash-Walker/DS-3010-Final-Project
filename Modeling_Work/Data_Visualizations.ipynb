```python
# Imports
import numpy as np
import pandas as pd
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Import data
x_train = pd.DataFrame(np.load("x_train.npy"))
y_train = pd.read_csv("y_train.csv", usecols = ["outcome"])

# Combine data
train_data = pd.concat([x_train, y_train], axis=1)

train_data.head()
```

```python
# Apply PCA
pca = PCA(n_components = 2)
pca_results = pca.fit_transform(train_data[[i for i in train_data.columns if i != "outcome"]])

# Visualize PCA
scatter_plot = plt.scatter(pca_results[:, 0], pca_results[:, 1], alpha = 0.6, c = train_data["outcome"].astype("category").cat.codes)
plt.legend(handles = scatter_plot.legend_elements()[0], labels = list(train_data["outcome"].astype("category").cat.categories))
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("PCA Analysis")
```
